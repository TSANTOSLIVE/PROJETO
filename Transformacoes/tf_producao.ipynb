```python
import pandas as pd
import re
```

```python
# lendo um arquivo csv, não existe nome de coluna
df = pd.read_csv('../Data/input/Producao/Producao.csv', sep=';')
```

```python
# drop column Id
df.drop(columns=['id'], inplace=True)
```

```python
# renomear coluna produto para Produto
df.rename(columns={'produto': 'Produto'}, inplace=True)
```

```python
# Função para substituir o valor da coluna que contém underline
def retorna_prefixo(linha_anterior):
    nova_palavra = ''
    palavra_maiuscula = ''
    if linha_anterior.isupper():  # Verificar se a linha anterior é toda maiúscula
        palavra_maiuscula = linha_anterior
        palavras_maiusculas = re.findall(r'[A-Z]+(?=\s|$)', linha_anterior)             
        for palavra in palavras_maiusculas:            
            nova_palavra = nova_palavra + palavra[0]
    return nova_palavra, palavra_maiuscula
```

```python
df['Codigo do Produto'] = ""
```

```python
# Inicializar a variável para armazenar 
linha_anterior = ''
prefixo = ''
palavra_maiuscula = ''

# Aplicar a função em cada linha do DataFrame
for index, row in df.iterrows():
    novo_prefixo, nova_palavra_maiuscula = retorna_prefixo(linha_anterior)
    if novo_prefixo:
        prefixo = novo_prefixo.lower()
        palavra_maiuscula = nova_palavra_maiuscula.lower()
    
    df.at[index, 'Codigo do Produto'] = prefixo + "_"+ row['Produto'].lower()
    # criar coluna tipo com valor de palavra_maiuscula
    df.at[index, 'Tipo'] = palavra_maiuscula
    linha_anterior = row['Produto']

    # Se a linha for toda maiscula dropa a linha
    if row['Produto'].isupper():
        df.drop(index, inplace=True)
```

```python
# Ordernar colunas, sendo que as ultimas colunas representam anos 2022 para trás, deve ser populado com for
columns = ['Codigo do Produto', 'Produto', 'Tipo']
n = len(df.columns) - len(columns)
ano_limite = 2022 + 1
ano_inicial = ano_limite - n


# Ordenar as colunas
df = df[columns + [str(i) for i in range(ano_inicial, ano_limite)]]
```

```python
#Salvando o arquivo
df.to_csv('../Data/output/Producao.csv', index=False, encoding='utf-8')
```
